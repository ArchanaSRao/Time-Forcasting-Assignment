# FORECASTING the DATA using Vector Auto Regression.

```python
Train = data_cleaned1.head(147)
Test = data_cleaned1.tail(12)
```

```python
nobs = 12
train = data_cleaned1[:-nobs]
test = data_cleaned1[-nobs:]
```

```python
len(train), len(test)
```

```python
train.info()
```

```python
p = [1,2,3,4,5,6,7]  # try with list of 7 p values

for i in p:
    model = VAR(train)
    results = model.fit(i)
    print(f'VAR Order {i}')
    print('AIC {}'.format(results.aic))
    print('BIC {}'.format(results.bic))
    print()
```

```python
model.endog_names
```

```python
results = model.fit(5)
results.summary()
```

```python
lag_order = results.k_ar
lag_order
```

```python
z = results.forecast(y=train.values[-lag_order:],steps = 12)
z
```

```python
test
```
